```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report

# Load the dataset
hotel_df = pd.read_csv('merged_dataset.csv')  # Replace with your dataset path

# Fill missing values for numeric columns
hotel_df['lead_time'].fillna(hotel_df['lead_time'].median(), inplace=True)
hotel_df['total_guests'].fillna(hotel_df['total_guests'].median(), inplace=True)
hotel_df['stays_in_weekend_nights'].fillna(0, inplace=True)
hotel_df['stays_in_week_nights'].fillna(0, inplace=True)
hotel_df['previous_cancellations'].fillna(0, inplace=True)
hotel_df['previous_bookings_not_canceled'].fillna(0, inplace=True)
hotel_df['adr'].fillna(hotel_df['adr'].median(), inplace=True)
hotel_df['required_car_parking_spaces'].fillna(0, inplace=True)
hotel_df['total_of_special_requests'].fillna(0, inplace=True)

# Handle categorical columns
hotel_df['market_segment'] = hotel_df['market_segment'].astype('category')
hotel_df['market_segment'] = hotel_df['market_segment'].cat.add_categories('unknown')
hotel_df['market_segment'].fillna('unknown', inplace=True)

hotel_df['day_of_week'] = hotel_df['day_of_week'].astype('category')
hotel_df['day_of_week'] = hotel_df['day_of_week'].cat.add_categories('unknown')
hotel_df['day_of_week'].fillna('unknown', inplace=True)

# Encode categorical variables
label_enc = LabelEncoder()
hotel_df['market_segment'] = label_enc.fit_transform(hotel_df['market_segment'])
hotel_df['day_of_week'] = label_enc.fit_transform(hotel_df['day_of_week'])

# Define the target variable (is_canceled) and features
features = [
    'lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights',
    'adults', 'children', 'total_guests', 'market_segment',
    'is_repeated_guest', 'previous_cancellations', 'previous_bookings_not_canceled',
    'adr', 'required_car_parking_spaces', 'total_of_special_requests', 'day_of_week'
]
X = hotel_df[features]
y = hotel_df['is_canceled']

# Handle any remaining missing values
X.fillna(0, inplace=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)

# Scale the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Logistic Regression Model
model = LogisticRegression(random_state=23, solver='liblinear')
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluate the model
acc_score = accuracy_score(y_test, y_pred)
pre_score = precision_score(y_test, y_pred)
rec_score = recall_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Print results
print(f'Accuracy Score: {acc_score}')
print(f'Precision Score: {pre_score}')
print(f'Recall Score: {rec_score}')
print('\nClassification Report:')
print(class_report)
```
